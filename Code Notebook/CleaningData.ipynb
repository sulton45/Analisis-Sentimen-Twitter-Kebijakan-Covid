```python
def load_data():
    data = pd.read_csv('CrawlingGabunganOKT.csv')
    return data

tweet_df = load_data()
tweet_df.tail(10)
```

```python
# definisi dataframe
df  = pd.DataFrame(tweet_df[['Date', 'User', 'Tweet']])

df.head(10)
```

```python
df.count()
```

```python
# Cleansing
def cleansing(tweet):
    # remove old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    # Remove additional code
    tweet = tweet.replace("\\xe2\\x80\\xa6", "")
    # Convert www.* or https?://* to URL
    tweet = re.sub("((www\.[^\s]+)|(https?://[^\s]+))", "", tweet)
    # Convert @username to AT_USER
    tweet = re.sub("@[^\s]+", "", tweet)
    # Remove additional white spaces
    tweet = re.sub("[\s]+", " ", tweet)
    # Replace #word with word
    tweet = re.sub(r"#([^\s]+)", r"\1", tweet)
    # Hapus angka
    tweet = re.sub(r"\d+", "", tweet)
    # Ganti tanda baca dengan spasi
    tweet = tweet.translate(str.maketrans(string.punctuation, " " * len(string.punctuation)))
    return tweet
df['Tweet Cleansing'] = df['Tweet'].apply(lambda x: cleansing(x))
```

```python
df.tail(50)
```

```python
def casefolding(tweet):
    # lower case
    tweet = tweet.lower()
    return tweet
df['Tweet Case Folding'] = df['Tweet Cleansing'].apply(casefolding)
```

```python
df.tail(50)
```

```python
from nltk.tokenize import word_tokenize 
# NLTK Word Tokenize 
def word_tokenize_wrapper(text):
    return word_tokenize(text)

df['Tweet Tokenize'] = df['Tweet Case Folding'].apply(word_tokenize_wrapper)
```

```python
df.tail(50)
```

```python
# NLTK calc frequency distribution / Cek frekuensi Tiap Kata
from nltk.probability import FreqDist
def freqDist_wrapper(text):
    return FreqDist(text)

df['Frekuensi Tweet'] = df['Tweet Tokenize'].apply(freqDist_wrapper)

print(df['Frekuensi Tweet'].head().apply(lambda x : x.most_common()))
```

```python
df.tail(50)
```

```python
# Menghapus Stopword
from nltk.corpus import stopwords

# NLTK stopword
# stopword indonesia
list_stopwords = stopwords.words('indonesian')


# Tambah Manual Stopword
list_stopwords.extend(["yg", "dg", "rt", "dgn", "ny", "d", 'klo', 
                       'kalo', 'amp', 'biar', 'bikin', 'bilang', 
                       'gak', 'ga', 'krn', 'nya', 'nih', 'sih', 
                       'si', 'tau', 'tdk', 'tuh', 'utk', 'ya', 
                       'jd', 'jgn', 'sdh', 'aja', 'n', 't', 
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt',
                       '&amp', 'yah'])

# Tambah Stopword dari txt
txt_stopword = pd.read_csv("stopwords.txt", names= ["stopwords"], header = None)

list_stopwords.extend(txt_stopword["stopwords"][0].split(' '))


list_stopwords = set(list_stopwords)


#hapus stopword di list tokenize
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

df['Tweet Filtering'] = df['Tweet Tokenize'].apply(stopwords_removal) 
```

```python
df.tail(50)
```

```python
# Normalisasi Kata
normalizad_word = pd.read_csv("normalisasi.csv")

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

df['Tweet Normalisasi'] = df['Tweet Filtering'].apply(normalized_term)

df['Tweet Normalisasi'].head(10)
```

```python
df.head(50)
```

```python
# Steaming Kata
# import Sastrawi package
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter


# buate stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in df['Tweet Normalisasi']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")


# apply stemmed term ke dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df['Tweet Stemmed'] = df['Tweet Normalisasi'].swifter.apply(get_stemmed_term)
print(df['Tweet Stemmed'])

# Pembersihan Terakhir untuk tanda petik dan kurung
def remove_punct(text):
    text  = " ".join([char for char in text if char not in string.punctuation])
    return text

df['Tweet Bersih'] = df['Tweet Stemmed'].apply(lambda x: remove_punct(x))
```
